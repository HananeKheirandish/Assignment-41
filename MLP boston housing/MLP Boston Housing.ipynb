```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
import numpy as np
import wandb
from wandb.keras import WandbCallback
```

```python
wandb.init(project='Boston Housing')
```

```python
config = wandb.config
config.learning_rate = 0.001
```

### Data

```python
dataset = tf.keras.datasets.boston_housing
(X_train, Y_train), (X_test, Y_test) = dataset.load_data()
```

### Model

```python
model = tf.keras.models.Sequential([
    Dense(24, input_dim=13, activation='sigmoid'), # HL1
    Dense(8, activation='sigmoid'), # HL2
    Dense(4, activation='sigmoid'), # HL3
    Dense(1, activation='linear') # OL
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate= config.learning_rate),
             loss=tf.keras.losses.MeanSquaredError())
```

```python
model.summary()
```

### Train

```python
model.fit(X_train, Y_train, epochs=320, callbacks=[WandbCallback()])
```

```python
model.evaluate(X_test, Y_test)
```

### Inference

```python
my_house = np.array([[0.00632, 18.0, 2.31, 0.0, 0.538, 6.575, 65.2, 4.09, 1.0, 296.0, 15.3, 396.9, 4.98]])

model.predict(my_house)
```
